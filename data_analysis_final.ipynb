```python
# Clean column names for easier handling
df_clean = df.copy()
df_clean.columns = ['S_No', 'Economic_Activity', 'NIC_Code', 'Expenditure_2018_19', 'Expenditure_2019_20', 'Expenditure_2020_21']

df_final = df_clean.head(20)
```

```python
df_final
```

```python
# Create Total_Expenditure column - sum of expenditure across all years for each sector
expenditure_columns = ['Expenditure_2018_19', 'Expenditure_2019_20', 'Expenditure_2020_21']
df_clean['Total_Expenditure'] = df_clean[expenditure_columns].sum(axis=1)

# Display top 10 sectors by total expenditure
print("Top 10 Economic Activities by Total Expenditure (2018-2021):")
top_sectors = df_clean.nlargest(10, 'Total_Expenditure')[['Economic_Activity', 'Total_Expenditure']]
for idx, (col, row) in enumerate(top_sectors.iterrows(), 1):
    print(f"{idx:2d}. {row['Economic_Activity']}: ₹{row['Total_Expenditure']:.2f} crores")
```

```python
df_final['Total_Expenditure'] = df_final[expenditure_columns].sum(axis=1)
```

### Initial Data Analysis

```python
# Total expenditure in each year

# 2018-19
total_2018_19 = df_final['Expenditure_2018_19'].sum()
print(f"Total Expenditure in 2018-19: ₹{total_2018_19:.2f} crores")

# 2019-20
total_2019_20 = df_final['Expenditure_2019_20'].sum()
print(f"Total Expenditure in 2019-20: ₹{total_2019_20:.2f} crores")

# 2020-21
total_2020_21 = df_final['Expenditure_2020_21'].sum()
print(f"Total Expenditure in 2020-21: ₹{total_2020_21:.2f} crores")

# Total expenditure across all years
total = df_final['Total_Expenditure'].sum()
print(f"Total Expenditure (2018-2021): ₹{total:.2f} crores")

```

```python
# Increase by percentage

# from 2018-19 to 2019-20
increase_2018_19_to_2019_20 = ((total_2019_20 - total_2018_19) / total_2018_19) * 100
print(f"Increase from 2018-19 to 2019-20: {increase_2018_19_to_2019_20:.2f}%")
# from 2019-20 to 2020-21
increase_2019_20_to_2020_21 = ((total_2020_21 - total_2019_20) / total_2019_20) * 100
print(f"Increase from 2019-20 to 2020-21: {increase_2019_20_to_2020_21:.2f}%")
# from 2018-19 to 2020-21   
increase_2018_19_to_2020_21 = ((total_2020_21 - total_2018_19) / total_2018_19) * 100
print(f"Increase from 2018-19 to 2020-21: {increase_2018_19_to_2020_21:.2f}%")
# from 2018-19 to 2020-21
increase_2018_19_to_2020_21 = ((total_2020_21 - total_2018_19) / total_2018_19) * 100
print(f"Increase from 2018-19 to 2020-21: {increase_2018_19_to_2020_21:.2f}%")
```

### 1. Bar Plots - top 10 economic activities with highest R&D expenditure across years

```python
# Expenditure by Economic Activity for each year
plt.figure(figsize=(27, 9))
years = ['2018-19', '2019-20', '2020-21']
expenditure_columns = ['Expenditure_2018_19', 'Expenditure_2019_20', 'Expenditure_2020_21']
colors = ['skyblue', 'lightgreen', 'lightcoral']

for i, (year, col) in enumerate(zip(years, expenditure_columns)):
    plt.subplot(1, 3, i+1)
    
    # Sort by expenditure
    data_sorted = df_clean.sort_values(col, ascending=False).head(11)
    
    bars = plt.bar(range(len(data_sorted)), data_sorted[col], 
                   color=colors[i], alpha=0.7)
    plt.title(f'Top 10 R&D Expenditure by Economic Activity ({year})', fontweight='bold')
    plt.ylabel('Expenditure (in crores)')
    plt.xticks(range(len(data_sorted)), 
               [activity[:20] + '...' if len(activity) > 20 else activity 
                for activity in data_sorted['Economic_Activity']], 
               rotation=45, ha='right')
    
    

plt.tight_layout()
plt.savefig("top 10 economic activities with highest R&D expenditure across year.png")
plt.show()

```

### 3. Box Plots - Distribution analysis

##### a. R&D Expenditure Distribution Across All Years

```python
plt.figure(figsize=(8, 6))
data_for_boxplot = [df_final['Expenditure_2018_19'], df_final['Expenditure_2019_20'], df_final['Expenditure_2020_21']]
plt.boxplot(data_for_boxplot, labels=['2018-19', '2019-20', '2020-21'])
plt.ylabel('Expenditure (in crores)')
plt.title('R&D Expenditure Distribution Across Years')
plt.grid(True, alpha=0.3)
plt.savefig('boxplot_expenditure_all_years.png')
plt.show()
```

```python
# Find the outliers in each year's expenditure using IQR method

# 2018-19
Q1_2018 = df_final['Expenditure_2018_19'].quantile(0.25)
Q3_2018 = df_final['Expenditure_2018_19'].quantile(0.75)
IQR_2018 = Q3_2018 - Q1_2018
lower_bound_2018 = Q1_2018 - 1.5 * IQR_2018
upper_bound_2018 = Q3_2018 + 1.5 * IQR_2018
outliers_2018 = df_final[(df_final['Expenditure_2018_19'] < lower_bound_2018) | (df_final['Expenditure_2018_19'] > upper_bound_2018)]
print("Outliers in 2018-19 Expenditure:")
print(outliers_2018[['Economic_Activity', 'Expenditure_2018_19']])

# 2019-20
Q1_2019 = df_final['Expenditure_2019_20'].quantile(0.25)
Q3_2019 = df_final['Expenditure_2019_20'].quantile(0.75)
IQR_2019 = Q3_2019 - Q1_2019
lower_bound_2019 = Q1_2019 - 1.5 * IQR_2019
upper_bound_2019 = Q3_2019 + 1.5 * IQR_2019
outliers_2019 = df_final[(df_final['Expenditure_2019_20'] < lower_bound_2019) | (df_final['Expenditure_2019_20'] > upper_bound_2019)]
print("Outliers in 2019-20 Expenditure:")
print(outliers_2019[['Economic_Activity', 'Expenditure_2019_20']])

# 2020-21
Q1_2020 = df_final['Expenditure_2020_21'].quantile(0.25)
Q3_2020 = df_final['Expenditure_2020_21'].quantile(0.75)
IQR_2020 = Q3_2020 - Q1_2020
lower_bound_2020 = Q1_2020 - 1.5 * IQR_2020
upper_bound_2020 = Q3_2020 + 1.5 * IQR_2020
outliers_2020 = df_final[(df_final['Expenditure_2020_21'] < lower_bound_2020) | (df_final['Expenditure_2020_21'] > upper_bound_2020)]
print("Outliers in 2020-21 Expenditure:")
print(outliers_2020[['Economic_Activity', 'Expenditure_2020_21']])

```
